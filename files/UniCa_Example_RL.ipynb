## How to run simulations

Komondor can be run by calling its compiled main file ("komondor_main") and indicating the value for different flags.

Next, you will see two different ways of running Komondor, depending on whether MAB Agents are instantiated or not.

In any case, **it is important to know** that:


*   Any input files used by the simulator will be located in the `/input` folder.
*   Any output files generated by the simulator will be located in the `/output` folder.

Apart from that, Komondor generates a set of console logs that give meaningful information.


### 1) Simulations without agents

To simulate specific scenarios (described using "input node" files) **without agents**, one must indicate the following flags:

```python
# DEFINE SIMULATION ARGUMENTS (NO AGENTS)

# Path of the file that describes the Wi-Fi nodes
INPUT_FILE_NODES='../input/input_nodes.csv'

# Path of the file where output logs are going to be written
OUTPUT_FILE_LOGS='../output/logs_test.txt'

# Path of the file where output logs are going to be written
SIMULATION_CODE='AAX-LAB1'

# Flag to indicate whether "NODE LOGS" will be written into output files (1) or not (0)
FLAG_SAVE_NODE_LOGS=1

# Flag to indicate whether "SYSTEM LOGS" will be printed in the console (1) or not (0)
FLAG_PRINT_SYSTEM_LOGS=1

# Flag to indicate whether "NODE LOGS" will be printed in the console (1) or not (0)
FLAG_PRINT_NODE_LOGS=1

# Total simulated time (in seconds)
SIM_TIME=100

# Value of the random seed that initializes the random number generator
SEED=7
```

### Exercise 5:

Run a simulation to test your action-selection strategy. Compare it to ɛ-greedy (Exercise 3) by plotting the temporary throughput obtained by each of the mechanisms.

```python
# (response to exercise 5)

def run_simulation(input_agents_file, sim_code, seed):
    input_nodes = os.path.join(mypath, 'input/input_nodes.csv')
    input_agents = os.path.join(mypath, f'input/{input_agents_file}')
    output_logs = os.path.join(mypath, f'output/logs_{sim_code}_seed{seed}.txt')

    flags = {
        'save_node_logs': 1,
        'save_agent_logs': 1,
        'print_system_logs': 0,
        'print_node_logs': 0,
        'print_agent_logs': 0,
    }
    sim_time = 100

    cmd = f"cd {mypath}/main; ./komondor_main {input_nodes} {input_agents} {output_logs} {sim_code} {flags['save_node_logs']} {flags['save_agent_logs']} {flags['print_system_logs']} {flags['print_node_logs']} {flags['print_agent_logs']} {sim_time} {seed}"
    print(f"Ejecutando simulacion: {cmd}")
    os.system(cmd)

# Ejecutar 5 simulaciones para cada estrategia
N = 5
for seed in range(1, N+1):
    run_simulation('agents_egreedy.csv', 'AAX-EGREEDY', seed)
    run_simulation('agents_ml4net.csv', 'AAX-ML4NET', seed)

import re
import numpy as np
import matplotlib.pyplot as plt

def extract_throughput(log_file):
    times = []
    throughputs = []
    with open(log_file, 'r') as f:
        for line in f:
            if 'Average throughput' in line:
                try:
                    parts = line.strip().split(';')
                    time = float(parts[0])
                    match = re.search(r'Average throughput = ([\d.]+)', line)
                    if match:
                        throughput = float(match.group(1))
                        times.append(time)
                        throughputs.append(throughput)
                except:
                    continue
    return times, throughputs

def average_throughput_over_seeds(sim_code, N):
    all_times = []
    all_throughputs = []

    for seed in range(1, N+1):
        log_path = os.path.join(mypath, f'output/logs_{sim_code}_seed{seed}.txt')
        times, throughput = extract_throughput(log_path)
        all_times.append(times)
        all_throughputs.append(throughput)

    # Suponemos que los tiempos son iguales en todas las simulaciones o hacemos interpolación
    # Para simplificar, tomamos tiempos de la primera simulación
    ref_times = all_times[0]
    # Convertimos en matriz
    throughputs_matrix = []
    for th in all_throughputs:
        # Aseguramos igual longitud (o truncamos)
        length = min(len(ref_times), len(th))
        throughputs_matrix.append(th[:length])

    throughputs_matrix = np.array(throughputs_matrix)
    avg_throughput = np.mean(throughputs_matrix, axis=0)
    return ref_times[:length], avg_throughput

times_egreedy, avg_egreedy = average_throughput_over_seeds('AAX-EGREEDY', N)
times_ml4net, avg_ml4net = average_throughput_over_seeds('AAX-ML4NET', N)

plt.figure(figsize=(10, 5))
plt.plot(times_egreedy, avg_egreedy, label='ɛ-greedy')
plt.plot(times_ml4net, avg_ml4net, label='ML4Net')
plt.title('Throughput Promedio Temporal Comparativo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Throughput Promedio (Mbps)')
plt.legend()
plt.grid(True)
plt.show()
```
